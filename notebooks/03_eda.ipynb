```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Only load the columns you actually need
cols_to_use = ['STATION ID', 'TIME', 'AVAILABLE_BIKES', 'BIKE_STANDS', 'occupancy_rate']

bikes = pd.read_csv('C:\\my-dublin-bikes-project\\outputs\\bikes_cleaned.csv', 
                    usecols=cols_to_use,
                    parse_dates=['TIME'],
                    nrows=2000000) # Still recommend using nrows for speed
```

```python
# Example Station Plot
sample_station = bikes['STATION ID'].iloc[0]
station_df = bikes[bikes['STATION ID'] == sample_station].set_index('TIME')
weekly = station_df['AVAILABLE_BIKES'].resample('W').mean()

plt.figure(figsize=(10, 5))
weekly.plot(title=f"Station {sample_station} Trends")
plt.show()
```

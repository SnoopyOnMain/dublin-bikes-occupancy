```python
import pandas as pd
import os
```

```python
# 1. Get the list of files from the 'data' folder
folder_path = "C:/my-dublin-bikes-project/data"
csv_files = []

for f in os.listdir(folder_path):
    # Only pick .csv files and ignore the metadata file 'dublin.csv'
    if f.endswith('.csv') and f != 'dublin.csv':
        csv_files.append(os.path.join(folder_path, f))

print(f"✅ Found {len(csv_files)} files. Standardizing and merging now...")
```

```python
folder_path = "C:/my-dublin-bikes-project/data"
output_file = 'bikes_raw_merged.csv'
chunk_size = 50000  # Only load 50,000 rows at a time

dtypes = {
    'STATION ID': 'int16', 'BIKE_STANDS': 'int16',
    'AVAILABLE_BIKE_STANDS': 'int16', 'AVAILABLE_BIKES': 'int16',
    'STATUS': 'category', 'ADDRESS': 'category', 'NAME': 'category',
    'LATITUDE': 'float32', 'LONGITUDE': 'float32'
}

standard_columns = [
    'STATION ID', 'TIME', 'LAST UPDATED', 'NAME', 'BIKE_STANDS',
    'AVAILABLE_BIKE_STANDS', 'AVAILABLE_BIKES', 'STATUS', 'ADDRESS', 
    'LATITUDE', 'LONGITUDE'
]

# Fresh start
if os.path.exists(output_file):
    os.remove(output_file)

csv_files = [f for f in os.listdir(folder_path) if f.endswith('.csv') and f != 'dublin.csv']
print(f"Merging {len(csv_files)} files using chunking...")

# 2. The Chunked Loop
is_first_chunk = True

for f in csv_files:
    full_path = os.path.join(folder_path, f)
    
    # We use 'chunksize' to read the file in small pieces
    for chunk in pd.read_csv(full_path, dtype=dtypes, parse_dates=['TIME', 'LAST UPDATED'], chunksize=chunk_size):
        
        # Apply standard columns to this chunk
        chunk.columns = standard_columns
        
        # Write this piece to the hard drive immediately
        if is_first_chunk:
            chunk.to_csv(output_file, index=False, mode='w')
            is_first_chunk = False
        else:
            chunk.to_csv(output_file, index=False, mode='a', header=False)
            
    print(f"✅ Finished processing: {f}")

print(f"Done! Final file saved as {output_file}")
```
